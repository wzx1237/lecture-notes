## Supervised Learning
training data set是sensor的input: a set $Σ$ of n-dimensional vectors. 而我们的label就是在这种情况下所需要的action.

**Thershold logic circuit (TLU)** is also called perceptron. (denote: $f$ is an boolean function.)

$$ f = \left\{\begin{aligned} 1, \quad &\sum x_{i} \omega_{i} > \theta \\ 0, \quad &\text{otherwise} \end{aligned}\right.$$

where $x_{i}$ is the input, $\omega_{i}$ is the weight, and $\theta$ is the thershold.

**Note** that **Not All** Boolean Functions can be implemented by TLU. Those **can** called linear separable functions.

例子：$x_{1}\oplus x_{2}$不是一个linear separable function. 我们可以使用一些feature engineering来分离它.

Consider:
$$\sum_{i = 1}^{n} x_{i}\omega_{i} \ge \theta \iff \sum_{i = 1}^{n} x_{i}\omega_{i} - \theta \ge 0 \\
\text{take: } x_{n + 1} = 1, \omega_{n + 1} = -\theta, \text{So: } \sum_{i = 1}^{n + 1} x_{i}\omega_{i} \ge 0 $$

This enable us **integrate the thershold into input vector**. (as $x_{n + 1}$, a fixed parameter).

## Error-Correction Procedure
Given $\vec{x}$ is the input vector, then the weight change rule is: $$\vec{W} = \vec{W} + \eta(T-O)\vec{x}$$
where $\vec{W}$ is the weight vector, $T$ is the label (target), $O$ is the actual output, and $\eta$ is the learning rate.

The exact number of steps needed depends on:
1. Initial weight values
2. Learning rate in weight updating rule
3. Order of presentation of training examples

In machine learning, two popular ways of updating weights are so called **gradient descent** and **stochastic gradient descent**(随机梯度下降).

## Design Action Function
任务：让小机器人贴边走。

使用state mechine来实现这一点。

具体实现：
\begin{matrix}
x_{1} &x_{1} & x_{2} \\
x_{3} &\text{You}   & x_{2} \\
x_{3} &x_{4} &x_{4} \end{matrix}

如图：如果表中两个$x_{1}$都有block, 则：$x_{1} = 1$. (对$x_{2}, x_{3}, x_{4}$同理)

Given the four features, we can specify a function of these features that will select the right action for boundary-following as follows:
1. if none of the four features equal to 1, then move **north** (it could be any direction).
2. if $x_1 = 1$ and $x_2 = 0$, then move **east**;
3. if $x_2 = 1$ and $x_3 = 0$, then move **south**;
4. if $x_3 = 1$ and $x_4 = 0$, then move **west**;
5. if $x_4 = 1$ and $x_1 = 0$, then move **north**.

## Boolean Algebra Representation

We can interupt the above description using boolean algebra:
1. if $\overline{x_{1}}\overline{x_{2}}\overline{x_{3}}\overline{x_{4}}$, then move $north$.
2. if $x_{1}\overline{x_{2}}$, then move $east$
3. if $x_{2}\overline{x_{3}}$, then move $south$
4. if $x_{3}\overline{x_{4}}$, then move $west$
5. if $x_{4}\overline{x_{1}}$, then move $north$


## Production System
One convenient representation for an action function is production systems. A production system is a sequence of productions, which are rules of the form:
$$c → a$$
meaning if condition $c$ is true, then do $a$. Here $a$ could be a primitive action, a call to a procedure, or a set of actions to be executed simultaneously.

例：
The following is a production system representation of the action function for our boundary-following robot:
$$ \begin{align}x_4x_1 &→ north,\\
 x_2x_3 &→ south,\\
 1 &→ north.\\
 x_3x_4 &→ west,\\
 x_1x_2 &→ east \end{align}$$

## Genetic Programming
注意：这里Genetic Programming指的不是C++中的GP，而是由mechine learning model一步步改进现有的skeleton的一个过程.

步骤：
1. Generate Population of **programs** (production system/decision tree)
2. fitness function
3. run programs and evaluate fitness
4. desired? not -> fit again; yes -> finish

## Fitness Function
例：boundary following robot

Given a program, and a starting position for the robot, we run it until it has carried out 60 steps, and then **count the number of cells next to the wall that are visited during these 60 steps**. (Max=32, Min=0.)

但是，starting position可能会影响结果。因为有些starting position可能会直接靠着墙。所以，我们要选择很多的starting position来避免这一点。

For a given program, we do **ten of these runs** with the robot starting at ten randomly chosen starting positions. The **total count** of the next-to-the-wall cels visited is taken **as the fitness of the program**. (Max = 320, Min=0.)
