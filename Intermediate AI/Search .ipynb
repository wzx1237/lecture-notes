#Search Problem
In general, a search problem consists of
1. A **set of states**.
2. An initial state.
3. A set of **deterministic actions** (or operators): if an action is executable in a state, then it will produce a unique next state.
4. A goal test: to decide whether the current state is goal state.
5. A **path cost function**. (例子：length of path)

##Problem-Solving Agents
Problem-solving agents are often **goal-directed**.

**Key ideas**:
To **determine what the best action is**, a problem-solving agent systematically **considers the expected outcomes** of different possible sequences of actions that lead to some goal state.

A systematic **search process is conducted in some representation space**. (就是所有的operation, state, goal...所组成的space)

步骤：
1. Goal and problem formulation
2. Search process
3. Action execution

###Remark:
通常来说，我们会**使用tree来表示representaion space**. Termination states作为leaf node, operation作为edge, initial state作为root.

##Breadth-First Search (广度优先搜索)
方法：强行搜索所有可能情况，然后返回depth最小的那一个solution. (当找到第一个solution之后，程序会立即停止并返回solution)

Properties:
1. 如果problem存在solution, 我们的strategy一定能找到它
2. 如果有多个solution, 我们一定会返回那个shortest-path solution
3. 这个strategy的complexity非常大

##Search Strategies
The **search strategy determines the order in which nodes in the search tree are expanded**. Different search strategies lead to different search trees.

**Four different criteria** that determine what the right search strategy is for a problem:
1. **Completeness**: Is it **guaranteed to find a solution if one exists**?
2. **Time complexity**: How long does it take to find a solution?
3. **Space complexity**: How much memory is needed?
4. **Optimality**: Does it **find the “best” solution** if several different solutions exist?

例子：

对Breadth-First Search来说，它是completeness的，因为它会保证找到solution (if exist); 它也是Optimality的，因为它一定会找到最优解

方法：和**inorder的顺序相同**，依次搜寻solution

Depth-first search generates the successors of a node one at a time; As soon as a successor is generated, one of its successors is generated;

##Avoiding Repeated States
There are **three ways to deal with repeated states**, in increasing order of effectiveness and computational overhead:
1. **Do not return to the state just came from**.
2. **Do not create a path with cycles**, that is, do not expand a node whose state has appeared in the path already.
3. **Do not generate any state that was generated before**.

注意到：第三种方法肯定能避免repeating, 但是它并不是十分effiency

#Heuristic Search
你需要一个**evaluation function (heuristic function)**, 用来衡量你现在的状态距离goal state相距多远。

例子：(8-puzzle game)
$$f(n) = \text{number of tiles out of place compared with goal}$$
这里，$f(n)$是你的evaluation function.
我们在决定下一步的时候要**一次性看多层的loss (当前状态的下一状态(next step)的loss, 和此前所有state的loss)**, 并选择这它们中loss最小的那个state作为新的"starting state"

We can add the path cost $g(n)$: number of edges we cover.

##A∗ Search
Evaluation function:
Sum of:
 1. actual path cost $g(n)$ from the start node to node n
 2. estimated cost $h(n)$

Estimated cost of the cheapest path through node n

Idea: Try to find the cheapest solution.

##A* Search by Tree
A∗ search by trees- check only ancestors for repeated states:
1. Create a search tree $T$ and a sorted list called $OPEN$, consisting solely of the start node, $n_0$. Put $n_0$ on $OPEN$
2. If $OPEN$ is empty, then exit with failure.
3. Select the first node in $OPEN$, and remove it from $OPEN$. Call this node $n$.
4. If $n$ is a goal node, exit successfully with the solution corresponding to the path from root $n_0$ to this node.
5. Expand node $n$, generate the set $M$ of its successors that are not already ancestors of $n$ in $T$. Install these members of $M$ as children of $n$ in $T$, and add them to $OPEN$.
6. Reorder the list $OPEN$ in order of increasing $g(n) + h(n)$ values. (Ties are resolved in favor of the deepest node in the search tree.)
7. Go to step 2.

核心思想：走向目前**可探索state**中cost最小的那一个。注意：这里**可探索state**指的是在我们的min-heap中存储的那些state。

##Analyse on A* search
We shall assume:
1. $h$ is **admissible**, that is, it is **never larger than the actual cost**.
2. $g$ is the **sum of the cost of the operators along the path**.
3. The cost of each operator is greater than some positive amount, $ϵ$.
4. The number of operators is finite (thus finite branching factor of search tree).

Under these conditions, we can always **revise $h$ into another admissible heuristic function so that the $f = h+g$ values along any path in the search tree is never decreasing**. If $f^{*}$ is the cost of an optimal solution, then:
1. A∗ expands all nodes with $f(n) < f^{*}$.
2. A∗ may expand some nodes with $f(n) = f^ {∗}$ before finding goal.

故，我们得到了如下结论：
1. A∗ is **complete**: A∗ expands nodes in order of increasing $f$, there are only finite number of nodes with $f(n) ≤ f^{∗}$, so it must eventually expanded to reach a goal state.
2. A∗ is **optimal**: it will always return an optimal goal.
3. It is **optimally efficient**: no other optimal algorithm is guaranteed to expand fewer nodes than A∗ (Dechter and Pearl 1985).



##Complexity of A* Search
Complexity:
1. Number of nodes expanded is exponential in the length of the solution.
2. All generated nodes are kept in memory. (A∗ usually runs out of space long before it runs out of time.)
3. With a good heuristic, significant savings are still possible compared to uninformed search methods.
4. Admissible heuristic functions that give higher values tend to make the search more efficient.

Memory-bounded extensions to A∗:
 Iterative deepening A∗ (IDA∗)
 Simplified memory-bounded A∗ (SMA∗)
