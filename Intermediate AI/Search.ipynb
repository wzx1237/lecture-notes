#Search Problem
In general, a search problem consists of
1. A **set of states**.
2. An initial state.
3. A set of **deterministic actions** (or operators): if an action is executable in a state, then it will produce a unique next state.
4. A goal test: to decide whether the current state is goal state.
5. A **path cost function**. (例子：length of path)

##Problem-Solving Agents
Problem-solving agents are often **goal-directed**.

**Key ideas**:
To **determine what the best action is**, a problem-solving agent systematically **considers the expected outcomes** of different possible sequences of actions that lead to some goal state.

A systematic **search process is conducted in some representation space**. (就是所有的operation, state, goal...所组成的space)

步骤：
1. Goal and problem formulation
2. Search process
3. Action execution

###Remark:
通常来说，我们会**使用tree来表示representaion space**. Termination states作为leaf node, operation作为edge, initial state作为root.

##Breadth-First Search (广度优先搜索)
方法：强行搜索所有可能情况，然后返回depth最小的那一个solution. (当找到第一个solution之后，程序会立即停止并返回solution)

Properties:
1. 如果problem存在solution, 我们的strategy一定能找到它
2. 如果有多个solution, 我们一定会返回那个shortest-path solution
3. 这个strategy的complexity非常大

##Search Strategies
The **search strategy determines the order in which nodes in the search tree are expanded**. Different search strategies lead to different search trees.

**Four different criteria** that determine what the right search strategy is for a problem:
1. **Completeness**: Is it **guaranteed to find a solution if one exists**?
2. **Time complexity**: How long does it take to find a solution?
3. **Space complexity**: How much memory is needed?
4. **Optimality**: Does it **find the “best” solution** if several different solutions exist?
