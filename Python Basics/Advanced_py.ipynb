```python
# 9月4号，复习python
# list comprehension
# 格式：
# <new-list-name> = [<expression> for <elements> in <list-name> if <condition>]
# 示例：
from math import sqrt
nums1 = [ sqrt(x) for x in range(1, 30) if (x % 4)]
print(nums1)
# output: 1.0, 1.414, 1.732, ...

# dictionary comprehension
# 格式：
# <new-dictionary-name> = {x : <expression> for <elements> in <list-name> if <condition>}
nums2 = [1, 2, 3, 4]
nums3 = {x: x**2 for x in nums2}
print(nums3)
# output:{1:1, 2: 4, 3: 9, 4: 16}

# set comprehension:
# 格式：
# <new-set-name> = {<expression> for <elements> in <list-name> if <condition>}
# 示例：
nums4 = {int(sqrt(i)) for i in range(30)}
print(nums4)
# output；{0, 1, 2, 3, 4, 5}


# tuple:
#  tuple is in many ways similar to a list; one of the most important
#  differences is that tuples can be used as keys in dictionaries and as elements
#  of sets, while lists cannot.
# example:
d = {(x, x + 1): x for x in range(10)} # Create a dictionary with tuple keys

t = (5, 6)          # Create a tuple
print(type(t))      # Print "<class 'tuple'>"
print(d[t])         # Print "5"
print(d[(1, 2)]) # Print "1"


# parallel iteration: 使用zip（）将多个list, set, ...连在一起
# 这样在进行loop的时候能joint together
# 示例：
fruits = ['apple', 'peach', 'banana', 'guava', 'papaya']
colors = ['red', 'pink', 'yellow', 'green', 'orange', 'black']
for name, color in zip(fruits, colors):
  print(name, 'is', color)
# output:
'''
apple is red
peach is pink
banana is yellow
guava is green
papaya is orange
'''
# 注意事项：由于color的长度更长，所以进行for loop的时候只会loop长度更短的那个
# 所以，black不会输出


# function: parameter passing
# 我们可以使用 * 来传递一堆arguments
# 示例：
def print_months(*months):
  print("The last month is " + months[-1])
# Print "The last month is December"
print_months("January", "Febuary", "March", "April", "May", "June",
"July", "August", "September", "October",
"November", "December")

# 需要注意的是：这里的argument都是以tuple形式传过去的，不能更改
# 如果使用 ** 来传递一堆arguments,会以dictionary的形式传过去

# 格式：
# def <function-name>(**<argument-name>):
#   ...(<argument-name>["<key>"]...)
# <function-name>(key = "...", ...)

# 示例：
def print_month(**month):
  print("The month is " + month["name"] + ", it is the " +
  month["order"] + "th month, it has " + month["days"] +
  " days.")
# Print "The month is September, it is the 9th month, it has 30 days."
print_month(name = "September", order = "9", days = "30")

# python里面也有protected和private的data
# 如何标注这些data？
# 在前面加上"_"是protect data，在前面加上"__"是private的data
# 但是实际上"__"在编译器处理的时候只是将这个变量重命名为
# _<class-name>__<variable-name>
# 它虽然仍然是一个protected data，但名字变了，这样子类就无法干扰父类
# 但是，一般来说这个不重要，因为即使非法访问也不会crush，只有warning罢了。
# 示例：
class Pokemon:
  # class variable: 相当于C++中的static data member
  pokemon_num = 0
  # 这个相当于constructor
  def __init__(self, name = 'Pikachu',
            index = '0025',
            types = 'electric'):
    # private instance variables
    self.__name = name
    self.__index = index
    self.__types = types
    Pokemon.pokemon_num += 1  # 每创建一个新的object,计数加一
  # accessor
  def print(self):
    print('---print Pokemon---')
    print('Name: '+self.__name)
    print('Index:'+self.__index)
    print('Types:'+self.__types)
  # mutator
  def set_name(self, name):
    self.__name = name
  def set__index(self, index):
    self.__index = index
  def set_types(self, types):
    self.__types = types

vulpix = Pokemon('Vulpix', '0037', 'Fire')
# error
# print('Name: ' + vulpix.__name)   # Attribute Error: ... object has no attribute ...
# print('Index: ' + vulpix.__index)
# print('Type: ' + vulpix.__types)
# Okay, but do not work
vulpix.__name = 'AlolanVulpix'
vulpix.__types = 'Ice'
# okay
vulpix.print()
togepi = Pokemon('Togepi', '0175', 'Fairy')
togepi.print()
print("Total Pokemon number is: ", Pokemon.pokemon_num)
```

```python
# 9月7号，复习python
# inheritance: 一种允许我们在已有的class上面搭建新的class的方式
# 所有的attribute和method都会在子类里面得到继承
# 格式：
'''
class <child-class-name> (<parent-class-name>):
  def __init__(self, <arg1>, <arg2>, ...):
    super().__init__(<parent-class-attributes>)
    self.<instance-variable1> = <value1>
    self.<instance-variable2> = <value2>
    ...
  def <function-name>(self, <parameter-list>):
    <function body>
'''
# 使用我们之前创建的class作为parent class
class Pokemon:
  # class variable: 相当于C++中的static data member
  pokemon_num = 0
  # 这个相当于constructor
  def __init__(self, name = 'Pikachu',
            index = '0025',
            types = 'electric'):
    # private instance variables
    self.__name = name
    self.__index = index
    self.__types = types
    Pokemon.pokemon_num += 1  # 每创建一个新的object,计数加一
  # accessor
  def print(self):
    print('---print Pokemon---')
    print('Name: '+self.__name)
    print('Index:'+self.__index)
    print('Types:'+self.__types)
  # mutator
  def set_name(self, name):
    self.__name = name
  def set__index(self, index):
    self.__index = index
  def set_types(self, types):
    self.__types = types

# 这是child class：
class Evol_Pokemon(Pokemon):
  def __init__(self, name = 'Pikachu', index = '0025', types = 'Electric',
            evol = 'Raichu'):
    # invoke parent class init:
    super().__init__(name, index, types)
    self.__evol = evol
  def set_evol(self, evol):
    self.__evol = evol
  def print(self):
    # invoke parent class的print:
    super().print()
    print('Evolve into: ', self.__evol)

jigglypuff = Evol_Pokemon('Jigglypuff', '0039', 'Normal', 'Wigglytuff')
jigglypuff.print()


# library:
# 格式：
# import特定东西
# from [module] import [function/class/variable]
# from [package].[module] import [function/class/variable]

# from [module] import * 这个表示import所有东西

# 还有import...as...
# 这个表示import进来的函数会换一种名称，一般是简写
# 例如：import numpy as np
#    import pandas as pd
# 如果使用from...import <name>
# 你可以直接写<name>调用这个函数
# 但是如果你使用import <package-name>.<module-name>.<function-name>
# 你在使用这个函数时，你需要写：<package-name>.<module-name>.<function-name>
```

```python
# 9月16日
# 如何计算高维array的乘法？
# 一般使用np.matmul来进行处理
# 运算规则：
# 1. 它将最后两个维度视为矩阵，而前面的所有维度视为批处理维度
# 2. broadcast:
# 如果两个array的shape分别是：(a, b, m, n)和(a, b, n, p)
# 那么他们可以直接相乘，如果shape是(a, 1, m, n)和(a, b, n, p)第一个array会在第二个维度上
# broadcast, 变成(a, b, m, n)再进行乘法
# 3. 矩阵乘法：
# 对于每一对批处理维度下的矩阵，独立的进行标准的矩阵乘法
# 例子：
import numpy as np
A = np.array(range(24)).reshape(2, 3, 4)
print(A)
# 这里我们创建了一个shape为2， 3， 4的array
```

```python
B = np.ones((2, 4, 5))
print(B)
```

```python
# 这里，我们发现A与B两个矩阵的最后两个维度满足矩阵乘法的要求可以相乘
# 我们需要的操作是：
# 对于批处理维度中的每一个i, 独立的计算A[i] @ B[i]
# 所以，我们的结果 C = A @ B
# 这里，C[0] = A[0] @ B[0]
#    C[1] = A[1] @ B[1]
# 所以，对于高维array的matmul运算：
# (..., m, n) @ (..., n, p) => (..., m, p)
# 本质上是一个批处理操作，只是将最后两个维度的矩阵按批乘在一起
C = A @ B
print(C)
print("C shape is: ", C.shape)
# 这里，C的shape是(2, 3, 5)
```

```python
# task：pairwise Distance
# 和score function有关
x = np.array([[ 8.54, 1.54, 8.12],
        [ 3.13, 8.76, 5.29],
        [ 7.73, 6.71, 1.31]])
y = np.array([[ 8.65, 0.27, 4.67],
        [ 7.73, 7.26, 1.95]])
reshaped_x = x.reshape(3, 1, 3)
reshaped_y = y.reshape(1, 2, 3)
diffs = reshaped_x- reshaped_y
# Output[i, j] represents the distance between x[i] and y[j].
print("Shape of diffs", diffs.shape)  # shape: (3, 2, 3)
print(diffs)
# It is important to see, via
# broadcasting, that diffs[i,j]
# stores x[i]- y[j]!
```

```python
dists = np.sqrt(np.sum(diffs**2, axis=2)) # 这里axis=2表示最后一个维度，横行
print(dists)
print(dists.shape)  # shape:(3, 2)
# 在我们最后的结果中，dists[i][j]表示x[i]和y[j]的距离
```
